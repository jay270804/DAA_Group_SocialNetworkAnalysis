```python
import networkx as nx
import random
import matplotlib.pyplot as plt

# Create an empty directed graph
G = nx.DiGraph()

# List of Indian first names
indian_names = [
    "Aarav", "Aarushi", "Advik", "Ananya", "Aryan", "Avani", "Dev", "Divya", "Ishaan", "Kavya",
    "Mira", "Neha", "Pranav", "Radha", "Rahul", "Riya", "Saanvi", "Sahil", "Samaira", "Samar",
    "Sanvi", "Shreya", "Siddharth", "Siya", "Soham", "Tara", "Ved", "Vivaan", "Yash",
    "Aadhya", "Aahana", "Aanya", "Aarohi", "Arnav", "Eesha", "Hritika", "Ishika", "Ishita", "Kritika",
    "Manya", "Myra", "Naira", "Navika", "Pari", "Maitri", "Riaan", "Rishaan", "Ruhan", "Rudra",
    "Shanaya", "Tia", "Vanya", "Vihaan"
]

# Number of nodes
num_nodes = 50

# Add nodes to the graph with Indian names
for name in indian_names:
    G.add_node(name)

# Add directed edges between nodes to represent following relationships
for i, name in enumerate(indian_names):
    # Randomly select the number of people the current person is following
    num_following = random.randint(0, 5)  # Maximum 5 people followed per person

    # Randomly select nodes to follow
    following = random.sample(indian_names[:i] + indian_names[i+1:], num_following)

    # Add edges from the current node to the nodes it is following
    for followed_name in following:
        G.add_edge(name, followed_name)

        # Add bidirectional edge with a 40% chance
        if random.random() < 0.4:
            G.add_edge(followed_name, name)
```

```python

communities = nx.community.greedy_modularity_communities(G)

# Assign community colors
node_color = {}
for i, community in enumerate(communities):
    for node in community:
        node_color[node] = i

# Define recommendation function
def recommend(user_id):
    similar_users = set(G.predecessors(user_id))
    recommended_users = []
    for user in similar_users:
        for followed_user in G.successors(user):
            if followed_user != user_id and not G.has_edge(user_id, followed_user):
                recommended_users.append(followed_user)
    return recommended_users


# Print communities with their corresponding node names and recommendations
for i, community in enumerate(communities):
    print(f"Community {i + 1}: {community}")

    # Recommendations within the community
    for user_id in community:
        recommendations = recommend(user_id)
        print(f"Recommendations for {user_id}: {recommendations}")

    # Draw the graph with communities highlighted
    pos = nx.kamada_kawai_layout(G.subgraph(community))
    colors = [node_color[node] for node in G.subgraph(community).nodes()]
    nx.draw(G.subgraph(community), pos, with_labels=True, node_size=700, node_color=colors, cmap=plt.cm.tab10, font_size=8, edge_color='gray', arrowsize=15)
    plt.title(f'Community {i + 1}')
    plt.show()
```

```python
# Draw the graph with communities highlighted
pos = nx.kamada_kawai_layout(G)
plt.figure(figsize=(15, 10))
colors = [node_color[node] for node in G.nodes()]
nx.draw(G, pos, with_labels=True, node_size=700, node_color=colors, cmap=plt.cm.tab10, font_size=8, edge_color='gray', arrowsize=15)
plt.title('Combined Graph with Communities')
plt.show()
```

```python
# Create bar graph
plt.figure(figsize=(12, 6))
plt.bar(indian_names, [len(list(G.predecessors(name))) for name in indian_names], color='skyblue')
plt.xlabel('User Names')
plt.ylabel('Number of Followers')
plt.title('Number of Followers for Each User')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
